# Real Estate Sales Analysis for 2001-2016

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Real_Estate_Sales_2001-2016.csv')
df.tail(10)
```

```python
import matplotlib.pyplot as plt
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
df02 = df.loc[df['ListYear']==2002]
df03 = df.loc[df['ListYear']==2003]
df16 = df.loc[df['ListYear']==2016]
df16.groupby(['Town'], as_index=False)['SaleAmount'].sum().nlargest(10, 'SaleAmount').plot.barh(x='Town')
plt.title('Top 10 Towns with highest Total Sales for the year 2016')
```

```python
df08 = df.loc[df['ListYear']==2008]
df08.groupby(['Town'], as_index=False)['SaleAmount'].sum().nlargest(10, 'SaleAmount').plot.barh(x='Town')
plt.title('Top 10 Towns with highest Total Sales for the year 2008')
```

```python
df02.groupby(['Town'], as_index=False)['SaleAmount'].sum().nlargest(10, 'SaleAmount').plot.barh(x='Town')
plt.title('Top 10 Towns with highest Total Sales for the year 2002')
```

*Inference: The towns Greenwich, Stamford, Fairfield and Westport retain their position in top 5 towns with highest total sales in all 3 years! Also, more-or-less the same names are seen in the list of top 10 with minor shuffle. This same analysis can be extended to account for top 50 towns as well, by changing the code to nlargest(50). Later we will see that most of this sales comes from Residential type Property and hence this can be linked to an indicator of town preference of people in general as well*

```python
df=df.drop(['ID'], axis=1)
df=df.drop(['SerialNumber'], axis=1)
```

```python
df.head()
```

```python
from pylab import rcParams
df_n1 = df.loc[df['PropertyType']=='Residential']
plt.subplot(1,2,1)
rcParams['figure.figsize'] = 15, 7
df_n1.groupby(['ListYear'])['SaleAmount'].mean().plot(kind='bar')
plt.ylabel('Sale Price in USD')
plt.title('Mean Sale Price Variation of Residential Property by year')
plt.subplot(1,2,2)
df_n1.groupby(['ListYear'])['PropertyType'].value_counts().plot(kind='barh')
plt.title('Number of Residential Property Sold by Year')
plt.xlabel('Number of Residential Property sold')
plt.subplots_adjust(wspace=0.7)
```

*The correlation (inverse relation) between Mean Residential Sale Price and Number of Residential Property Sales is not very intuitive as one might expect.
As can be seen, from 2006-12: we see a direct relation between the two; but after 2012, we start to see an inverse relation between the two. A strong relation outside these ranges cannot be easily derived. 
When we will see below, (First graph of In[36]), the highest total sales were in the years 2002, 2004 and 2016.
This supports the theory of 'Sweet spot of Pricing', we see highest sales (both number of, and value of) at or near a sweet spot, usually (but not necessarily) the median; as can be verified from the Mean Sale Price at 2002, 2004 and 2016. Also the effects of Recession is quite visible from both the graphs at and around 2007*

```python
from pylab import rcParams
df_n1 = df.loc[df['PropertyType']=='Residential']
plt.subplot(1,2,1)
rcParams['figure.figsize'] = 15, 7
df_n1.groupby(['ListYear'])['SaleAmount'].median().plot(kind='bar')
plt.ylabel('Sale Price in USD')
plt.title('Median Sale Price Variation of Residential Property by year')
plt.subplot(1,2,2)
df_n1.groupby(['ListYear'])['PropertyType'].value_counts().plot(kind='barh')
plt.title('Number of Residential Property Sold by Year')
plt.xlabel('Number of Residential Property sold')
plt.subplots_adjust(wspace=0.7)
```

```python
df_new = pd.read_csv('Affordable_Housing_by_Town_2011-Present.csv')
```

```python
plt.subplot(1,3,1)
rcParams['figure.figsize'] = 20, 7
df_n1.groupby(['ListYear'])['SaleAmount'].mean().plot(kind='bar')
plt.ylabel('Sale Price in USD')
plt.title('Mean Sale Price Variation of Residential Property by year')
plt.subplot(1,3,2)
df_new.groupby(['Year'])['Percent Affordable'].mean().plot(kind='line')
plt.ylabel('Mean Affordability(%)')
plt.title('Mean Affordability(%) v/s Year')
plt.subplot(1,3,3)
df_new['CHFA/ USDA Mortgages%'] = df_new[' Single Family CHFA/ USDA Mortgages']/df_new['2010 Census Units']*100
df_new.groupby(['Year'])['CHFA/ USDA Mortgages%'].mean().plot(kind='line')
plt.ylabel('CHFA/ USDA Mortgages Units as Percent of Total Units')
plt.title('Mean % CHFA/ USDA Mortgages Units with Year')
plt.subplots_adjust(wspace=0.2)
```

*Inference: Decrease in Mean Residential Sale Price leads to increase in Affordability. Also, Increase in CHFA/USDA Mortgage Assisted Units leads to increase in affordability, thereby increasing buying power of common man.
Another point to note is the low Mean Sale Price of 2015. Now, one might expect that with peak affordability in 2015, high CHFA as well as low Sale Price, the sales would have peaked. But this is not so (See In[36] First Graph). According to economists, this is because of low demand due to unattractive job market then.*

As regards the mean residential sale amount of two or more family residential types, one cannot rely on the figures, since most of the datapoints belong to single family residential type, although one can get a sense of the variation of figures across the types. The same can be said for Property type(Condo, Commercial and others) Besides, the effects of Recession and consequently the negative effect on Connecticut's economy can be inferred from the first two graphs*
